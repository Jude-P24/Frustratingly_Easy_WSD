```python
import pandas as pd
```

```python
OR_dic = pd.read_csv('OR_dict.csv')
```

```python
OR_dic
```

```python
df_filtered = OR_dic[OR_dic['word_count'] != 0]
df_filtered
```

```python
# Sort by 'full_spell' and 'odds_ratio' in descending order
df_grouped_sorted = df_filtered.sort_values(by=['full_spell', 'odds_ratio'], ascending=[True, False])
df_grouped_sorted.to_csv('OR1_WC_dict.csv', index=False)
```

```python
df_grouped_sorted
```

```python
df_top300 = df_grouped_sorted.groupby('full_spell').head(1)
df_top300.to_csv('OR1_WC_top1.csv', index=False)
```
